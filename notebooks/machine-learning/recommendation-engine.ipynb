```python
from textblob import Word
```

```python
lemma_corpus = [Word(words).lemmatize() for words in corpus]
```

```python
# remove caps, making all options lower_case
lower_case_corpus = [option.lower() for option in lemma_corpus]
```

```python
clean_corpus = [replacements.get(item, item) for item in lower_case_corpus]
```
