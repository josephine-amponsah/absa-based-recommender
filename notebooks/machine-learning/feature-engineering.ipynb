```python
corpus = []
for record in data['experience'].tolist():
    if len(record) != 0:
       corpus.extend(record['amenities'])
```

Create TF-IDF vector for amenities

```python
#preprocessing data
corpus = list(set(corpus))
```

```python
corpus
```

```python
from textblob import Word
```

```python
lemma_corpus = [Word(words).lemmatize() for words in corpus]
```

```python
lemma_corpus
```

```python
# remove caps, making all options lower_case
lower_case_corpus = [option.lower() for option in lemma_corpus]
```

```python
lower_case_corpus
```

```python
replacements = {
    'bar onsite': 'bar',
    'breakfast' : 'free breakfast',
    'kitchen in all rooms': 'kitchen in rooms',
    'kitchens in some rooms': 'kitchen in rooms',
    'parking' : 'free parking',
    'outdoor pool': 'pool',
    'fitness center': 'gym'
}
```

```python
clean_corpus = [replacements.get(item, item) for item in lower_case_corpus]
```

```python
clean_corpus
```

```python
preprocessed_corpus = list(set(lower_case_corpus))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
tf_idf_matrix = vectorizer.fit_transform(preprocessed_corpus)
```

```python
test_output = vectorizer.transform(['greens and mountains', 'beach', 'free breakfast'])
```

```python
test_output.toarray()
```

```python
def replace_options(record):
    if len(record) != 0:
        values = record['amenities']
        lemma_values = [Word(words).lemmatize() for words in values]
        lower_case_values = [option.lower() for option in lemma_values]
        record['amenities'] = [replacements.get(item, item) for item in lower_case_values]
    return record
```

```python
data['experience'] = data['experience'].apply(lambda x: replace_options(x))
```

```python
verify = []
for record in data['experience'].tolist():
    if len(record) != 0:
       verify.extend(record['amenities'])
```
