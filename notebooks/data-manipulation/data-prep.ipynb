```python
regions = {
    'Western Region': ['Takoradi', 'Asanta', 'Aboadze', 'Busua', 'Bogoso', 'Esiama', 'Busua', 'Akwidaa','Sekondi-Takoradi','Axim', 'Shama'],
    'Eastern Region': ['Akosombo', 'Awete3so', 'Senchi', 'Aburi','New Abirem','Atimpoku','Larteh', 'Akropong','Nsawam','Kyebi','Koforidua','Peduase',
                       'Nkwatia','Obo','Mamfe','Ashyresu', 'Anum','Adukrom','Akrade','Pepease','Peduasi','Old Tafo', 'Kuntanse'],
    'Central Region':['Cape Coast','Apam', 'Elmina','Gomoa Fetteh','Brenu','Sokode Etoe','Akrade','Adome','Kpong'],
    'Volta Region':['Sogakope','Elonyi','Hohoe','Abor','Ho','Biakpa','Keta','Vume','Akatsi','Dzita' , 'Betekpo', 'Aflao', 'Abonse', 'Tunu'],
    'Greater Accra': ['Ada','Accra', 'Ada Foah']
}
```

```python
def get_region(city):
    for region, cities_list in regions.items():
        if city in cities_list:
            return region
    return None

# Use the apply() function to create a new 'regions' column
hotel_facts['region'] = hotel_facts['city'].apply(get_region)
```
