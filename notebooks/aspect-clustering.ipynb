```python
from textblob import TextBlob,Word
```

```python
def label_similarity(labels, aspects, model):
    distance_dict = {}
    # aspect_list = [Word(word).lemmatize() for word in aspects]
    for aspect in aspects:
        word = Word(aspect).lemmatize()
        try:
            aspect_vec = model.wv[word]
            label_dist = {}
            for label in labels:
                try:
                    label_vec = model.wv[label]
                    similarity = similarity_cosine(aspect_vec, label_vec)
                    label_dist[label] = similarity
                except KeyError:
                    label_dist[label] = None
            distance_dict[aspect] = label_dist
        except KeyError:
            distance_dict[aspect] = None
    return distance_dict
```

```python
distance_data = label_similarity(labels, aspects, model)
```

```python
distance_data
```

```python
aspect_labels = {}
for record, values in distance_data.items():
    if values == None:
        aspect_labels[record] = None
    else:
        scores = []
        keys = []
        for key, score in values.items():
            keys.append(key)
            scores.append(score)
        max_index = scores.index(max(scores))
        aspect_labels[record] = keys[max_index]
        
    
```
